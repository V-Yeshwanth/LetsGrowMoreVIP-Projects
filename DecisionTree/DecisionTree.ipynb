```python
# Splitting of data
from sklearn.model_selection import train_test_split
Predictors=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X=df[Predictors]
y=df.Species
print(X.head())
print(y.head())
X_train,X_val,y_train,y_val = train_test_split(X, y, random_state=0)
```

```python
#Scatter plot
import seaborn as sns
plt.figure(figsize=(10,10))
sns.scatterplot(data=X,alpha=0.9)
plt.title("Data_Attributes",size=15)
plt.show()
```

```python
#correlation map
plt.figure(figsize=(7,7))
hm=X.corr()
sns.heatmap(hm,annot=True)
plt.title("Correlation map")
```

### **Defining Model**

```python
# Defining the model
from sklearn.tree import DecisionTreeClassifier
Tree=DecisionTreeClassifier(criterion='entropy',max_depth=4)
Tree.fit(X_train,y_train)
```

### **Prediction**

```python
Predictions=Tree.predict(X_val)
pd.DataFrame({'Actual':y_val,'Predicted':Predictions})
```

```python
from sklearn import metrics
print('Accuracy:',metrics.accuracy_score(y_val,Predictions))
```

```python
metrics.confusion_matrix(y_val,Predictions)
```
