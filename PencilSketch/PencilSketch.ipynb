### Import Required Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

### Read the Image

```python
image=cv2.imread("/Ironman-1.jpg")
cv2_imshow(image)
```

## STEP 1 Convert the RGB color image to grayscale

```python
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

## STEP 2 Invert the grayscale image to get a negative

```python
img_invert=cv2.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

## STEP 3 Apply a Gaussian blur to the negative from step 2

```python
img_smooth=cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
```

```python
cv2_imshow(img_smooth)
```

## STEP 4 Blend the grayscale image from step 1 with the blurred negative from step 3 using a color dodge

```python
def dodgeV2(x,y):
  return cv2.divide(x,255-y,scale=256)
```

```python
final_img=dodgeV2(img_gray,img_smooth)
```

```python
cv2_imshow(final_img)
```

# Final Output

```python
cv2_imshow(image)
cv2_imshow(final_img)
```
